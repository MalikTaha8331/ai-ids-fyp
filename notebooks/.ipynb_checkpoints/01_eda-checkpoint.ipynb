```python
from scipy.io import arff

# Load ARFF files
train_data, train_meta = arff.loadarff('../data/raw/KDDTrain2+.csv')
test_data, test_meta   = arff.loadarff('../data/raw/KDDTest1+.csv')

# Convert to pandas dataframe
train_df = pd.DataFrame(train_data)
test_df  = pd.DataFrame(test_data)

# Fix byte strings
for col in train_df.select_dtypes(include=['object']).columns:
    train_df[col] = train_df[col].str.decode('utf-8')
    test_df[col]  = test_df[col].str.decode('utf-8')

# Rename label column if needed
train_df.rename(columns={'class': 'label'}, inplace=True)
test_df.rename(columns={'class': 'label'},  inplace=True)

print("Train shape:", train_df.shape)
print("Test shape: ", test_df.shape)
train_df.head()
```
